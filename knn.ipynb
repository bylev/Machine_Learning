# K-NN Implementation

***K-NN intuition***

*K-nearest neighbots* (K-NN) is a classification algorithm based on the idea that points in a feature space are close to each other that tend to have similar labels.
To classify a point, we must find the "k" closest points based on some distance metrics and assign the most common label among these "k" neighbors to the point that we are classifying.

```python
import numpy as np
import matplotlib.pyplot as plt
```

3. Generate random data and define X (features) and Y (labels), where X is the average of kills per game and average of deaths per game, and Y is the takes 0 as ADC, 1 as tank and 2 as support.

```python
np.random.seed(0)
num_samples = 50 #* Number of samples

kills = np.random.uniform(1, 10, num_samples) #* Average kills
deaths = np.random.uniform(0, 5, num_samples) #* Average deaths
 
X= np.column_stack((kills, deaths)) #* Feature 
Y = np.random.randint(3, size=num_samples) #* Labels
```

4. Define the function calculate the distance between two points in the feature space, considering all the features available irrespecting of the number of labels.

```python
def euclidean_distance(x1, x2): #* Compute euclidean distance
    return np.sqrt(np.sum((x1 - x2) ** 2))  
```

5. Define the function to predict role using K-NN. 


```python
def predict_role(x_train, y_train, x_test, k):
    distances = np.array([euclidean_distance(x_test, x_train[i]) for i in range(num_samples)]) #* Compute distances to training data
    knn = np.argsort(distances)[:k] #* Get indexes
    near_labels = y_train[knn] #* Get labels
    unique_labels, counts = np.unique(near_labels, return_counts=True) #* Get unique labels
    return unique_labels[np.argmax(counts)] #* Return most common label
```

# CODE 

```python
import numpy as np
import matplotlib.pyplot as plt

#! Generate random data
np.random.seed(0)
num_samples = 50

kills = np.random.uniform(1, 10, num_samples)
deaths = np.random.uniform(0, 5, num_samples)

#* X: Average of kills per game and average of deaths per game
X = np.column_stack((kills, deaths))

#*Labels: 0 (ADC), 1 (Tank), 2 (Support)
Y = np.random.randint(3, size=num_samples)

#* Function to calculate the distance
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

#* Function to predict the role
def predict_role(x_train, y_train, x_test, k):
    distances = np.array([euclidean_distance(x_test, x_train[i]) for i in range(num_samples)])
    knn = np.argsort(distances)[:k]
    near_labels = y_train[knn]
    unique_labels, counts = np.unique(near_labels, return_counts=True)
    return unique_labels[np.argmax(counts)]

#* Function to plot
def plot(X, Y, k):
    plt.figure(figsize=(10, 6))
    colors = ['pink', 'purple', 'orange']
    markers = ['o', 'o', 'x']
    roles = ['ADC', 'Tank', 'Support']
    
#* Plot the data
    for i in range(3):
        plt.scatter(X[Y == i][:, 0], X[Y == i][:, 1], label=roles[i], color=colors[i], marker=markers[i], alpha=1,s=200)

    plt.xlabel('Average of kills per game')
    plt.ylabel('Average of deaths per game')
    plt.title(f'Role prediction using K-NN (k={k})')
    
    plt.legend()
    plt.grid(True)
    plt.show()
   

k = 5 #? number of neighbors
plot(X, Y, k)
k= 10
plot(X, Y, k)
k = 15 #? number of neighbors
plot(X, Y, k)
k= 20
plot(X, Y, k)
```

# Does K-NN have a loss function?

K-NN does not have a loss function that can be minimized during training. This algorithm is not trained at all. The only "training" that happens for it, is memorising the data (creating a local copy), so that during prediction you can do a search and majority vote. No function is fitted to the data, and so, no optimization is done (it cannot be trained using gradient descent). 

# Optimization Function

K-NN does not have an optimization function because it does not learn any model parameters. There are no weights or coefficients to optimize through an optimization algoritm as we mentioned before.
