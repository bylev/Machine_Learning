```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

file_path= r'C:\Users\marie\Downloads\Machine_Learning\indicators.csv'
df = pd.read_csv(file_path, encoding='latin-1')
df.head()


```

CLEAN DATASET

```python
df.dtypes
# Para contar valores nulos en cada columna
null_counts = df.isna().sum()
# print(null_counts)

#rellenar valores nulos
df.bfill(inplace=True)
print(null_counts)

```

```python
#Drop first five columns
df.drop(df.columns[0:5], axis=1, inplace=True)
df.head()
```

```python
# Convert into numerical
df = pd.get_dummies(df, dtype='int64')
df
```

```python
N_poverty= df['N_pobreza']
df= df.drop('N_pobreza', axis=1)
df['N_poverty']= N_poverty
df.head()
```

```python
df.isnull().sum()
```

```python
X= np.asarray(df.values.tolist())
Y= np.asarray(df['N_poverty'].values.tolist())
Y= Y.reshape(-1,1)

#fill nan values x
X= np.nan_to_num(X)
```

```python
print("The shape of the independent variables is: ", X.shape)
print("The shape of the dependent variable is: ", Y.shape)
```

```python
#Use 70% of the data for training and 30% for testing
X_train= np.asarray(X[:int(0.7*len(X))])
Y_train= np.asarray(Y[:int(0.7*len(Y))])
X_test=np.asarray(X[int(0.7*len(X)):])
Y_test= np.asarray(Y[int(0.7*len(Y)):])

X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create and fit the regression model
regression_model = LinearRegression()
regression_model.fit(X_train, Y_train)

# Make predictions on the test data
Y_predicttt = regression_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(Y_test, Y_predicttt)
r2 = r2_score(Y_test, Y_predicttt)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
print(Y_predicttt.shape)
```

```python
# Standardize the training set
X_train_standardized = (X_train - X_train.mean()) / X_train.std()

# Standardize the test set using mean and std from the training set
X_test_standardized = (X_test - X_train.mean()) / X_train.std()
```

```python
def error_function(y_actual, y_predicted):
    error = 0
    for i in range(0, len(y_actual)):
        error = error + pow((y_actual[i] - y_predicted[i]), 2)
    return error / (2 * len(y_actual))

def y_predicted(w, x):
    y_pred = np.zeros(len(x))
    for i in range(0, len(x)):
        for j in range(0, len(w) - 1):  # Exclude the last element in w
            y_pred[i] = y_pred[i] + (w[j] * x[i][j])
        y_pred[i] += w[-1]  # Intercept
    return y_pred
```

```python

def gradient_descent(y_actual, y_pred, x):
    grad = np.zeros(x.shape[1])
    for i in range(x.shape[1]):
        for j in range(0, len(y_actual)):
            grad[i] = - (y_actual[j] - y_pred[j]) * x[j][i] + grad[i]
    return grad / len(y_actual)

def weights(x_train, y_train, num_iterations, learning_rate):
    no_of_rows = x_train.shape[0]
    no_of_columns = x_train.shape[1]
    new_x_train = np.ones((no_of_rows, no_of_columns + 1))
    new_x_train[:, 0:no_of_columns] = x_train
    w = np.zeros(no_of_columns)
    w = np.append(w, 1)
    for i in range(0, num_iterations):
        y_pred = y_predicted(w, new_x_train)
        error = error_function(y_train, y_pred)
        print("mean square error: ", error, "after", i, "th iteration")
        MSE_points.append(error)
        grad = gradient_descent(y_train, y_pred, new_x_train)
        w = w - learning_rate * grad
        learning_rate = learning_rate / (1 + i * 0.1) # Learning rate Decay
    return w
```

```python

def regression_test(x_test, w):
    row = x_test.shape[0]
    column = x_test.shape[1]
    new_x_test = np.ones((row, column + 1))
    new_x_test[:, 0:column] = x_test
    y_pred = y_predicted(w, new_x_test)
    return y_pred

def r_squared(y_actual, y_predicted):
    y_mean = np.mean(y_actual)
    ss_tot = sum((y_actual - y_mean) ** 2)
    ss_res = sum((y_actual - y_predicted) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    return r2

MSE_points = []
num_iterations = 30
learning_rate = 0.0001
w = weights(X_train_standardized, Y_train, num_iterations, learning_rate)
Y_pred = regression_test(X_test_standardized, w)
mse = error_function(Y_test, Y_pred)
r2 = r_squared(Y_test, Y_pred)
print("R-squared on Test Set: ", r2)
print("Predicted Y values: ", Y_pred)
```
